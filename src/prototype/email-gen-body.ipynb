```python
import base64
from email.mime.text import MIMEText

def create_message_encoded(sender, to, subject, message_text):
    # 1. Create the RFC 2822 MIME message
    message = MIMEText(message_text)
    message['to'] = to
    message['from'] = sender
    message['subject'] = subject

    # 2. Convert message to bytes (as_bytes() ensures correct binary handling)
    raw_bytes = message.as_bytes()

    # 3. Encode as base64URL (RFC 4648)
    # urlsafe_b64encode handles '-' and '_' replacements
    encoded_message = base64.urlsafe_b64encode(raw_bytes).decode('utf-8')

    return {'raw': encoded_message}

# Example usage
email_data = create_message_encoded(
    sender="ches.cmu@gmail.com",
    to="nnnpooh@gmail.com",
    subject="Hello!",
    message_text="This is an RFC 2822 message encoded for API use."
)
print(email_data['raw'])
```
